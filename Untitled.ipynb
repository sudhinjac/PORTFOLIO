```python

import pandas as pd 
import yfinance as yf 
import pyfolio as pf

# setup list of tickers i.e. portfolio companies

tickers = ["TANLA.NS","HCLTECH.NS","ALKYLAMINE.NS"]
tickers

df = pd.DataFrame(columns=tickers)
df

# Fetch the data 

for ticker in tickers:
  df[ticker] = yf.download(ticker, period="2y")['Adj Close']

df

# 1. calculate the returns of individual stocks and then compute daily return
# 2. Mean return of the daily portfolio returns

df = df.pct_change().dropna().mean(axis=1)
df

pf.create_simple_tear_sheet(df)

```
